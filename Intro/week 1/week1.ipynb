#### Lecture 2: Doing More with Graphis to Model Data

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-3, 3, 20)
y = x ** 2
```

```python
plt.plot(x,y,'b-')
```

```python
plt.plot(x,y,'g*-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Our Very First Plot')
```

```python
plt.subplot(1,2,1) #One row, two columns, first graph
plt.plot(x,y,'g')
plt.title('Y is a function of X')

plt.subplot(1,2,2) #One row, two columns, second graph
plt.plot(y,x,'r--')
plt.title('X is a function of Y')
```

### Introduction to the Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.9, 0.8]) #Left, bottom, width, left
axes.plot(x,y, 'b*')
```

```python
fig = plt.figure(figsize=(10,8))

axe1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axe1.plot(x, y, 'b-o', lw=2)
axe1.set_title('Larger Graph')

axe2 = fig.add_axes([0.1, 0.1, 0.2, 0.3])
axe2.plot(y, x, 'r->', lw=2)
axe2.set_title('Smaller Graph')
```

### subplots()

The plt.subplots() object will act as a more automatic axis manager

Basic use cases:

```python
#Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

#Now use axes object to add stuff to plot
axes.plot(x,y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

Then you can specify the number of rows and columns when creating the subplots() object:

```python
#Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,5))
```

```python
axes #Shows that this object is simply an array
```

We can iterate through this array:

```python
for ax in axes:
    ax.plot(x,y,'m')
    ax.set_xlabel('x-label')
    ax.set_ylabel('y-label')
    ax.set_title('Repeating')

#Display the figure object
fig
```

## Saving figures
Matplotlib can generate high-1uality output in a number of formats, including png, jpeg, eps, pgf, and pdf

To save a figure to a file we can use the `savefig` method in the `Figure` class:

```python
fig.savefig('sample_figure.png')
#We can optionally specify the DPI and choose between output formats:
fig.savefig('sample_figure200DPI.png', dpi=200)
```

### Legends, lables, and titles

You can use `label='label text` keyword argument when plots or other objects are added to the figure, and then using the `legend` method without arguments to add the legend to the figure:

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, 'r*',  label="x**2")
ax.plot(x, x**3, 'g--', label="x**3")
ax.legend(loc=0)#loc=1: upper right; loc=2 upper left; loc=3 lower left; loc=4 lower right
```
